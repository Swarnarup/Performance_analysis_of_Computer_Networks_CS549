```python
import pandas as pd
import matplotlib.pyplot as plt
import math

sent = 1000000
```

```python
def p_B(ro, B):
    return (ro**B)*(1-ro)/(1-(ro**(B+1)))

def waitTime(arr, serv, ro, B):
    p0 = (1-ro)/(1-(ro**(B+1)))
    # N = (1/((1-ro)**2)) - (B*(ro**B)/(1-ro))
    N = ro*(((1-(ro**(B+1))) / ((1-ro)**2)) - ((B+1)*(ro**B)/(1-ro)))
    return (p0*N*arr) - serv

# def waitTimeMG1(arr, serv, sd, ro):
#     lb = 1/arr
#     mu = 1/serv
#     return (lb*(mu*mu)) / (2*(1-ro))
```

### M/G/1/4 (ro vs pktLoss & ro vs waitTime)

```python
df = pd.read_excel("mg13.xlsx")
```

```python
df = pd.read_excel("mg13.xlsx")

plt.title("pkt Loss % vs ro for M/G/1/4")
plt.plot(df["ro"], df["pktLoss"]/sent, label = "Simulated")
plt.scatter(df["ro"], df["pktLoss"]/sent, label = "data points")
plt.plot(df["ro"], [p_B(ro, 1) for ro in df["ro"]], label = "from calculation", color = 'green',  linestyle = 'dashed')
plt.legend()
plt.xlabel("ro (0.1 to 1.9) ->")
plt.ylabel("pkt loss %")
plt.show()

plt.title("avg queuing delay vs ro for M/G/1/4")
plt.plot(df["ro"], df["avgWaitTime"], label = "Simulated")
plt.scatter(df["ro"], df["avgWaitTime"], label = "data points")
plt.plot(df["ro"], [waitTime(df["avgArrTime"][i], df["avgServTime"][i], df["ro"][i], 1) for i in range(len(df["ro"]))], label = "from calculation", color = 'green',  linestyle = 'dashed')
plt.legend()
plt.xlabel("ro (0.1 to 1.9) ->")
plt.ylabel("Queuing dely (seconds)")
plt.show()
```

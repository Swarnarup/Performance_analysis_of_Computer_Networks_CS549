```python
import pandas as pd
import matplotlib.pyplot as plt

sent = 1000000
```

```python
def p_B(ro, B):
    return (ro**B)*(1-ro)/(1-(ro**(B+1)))

def waitTime(arr, serv, ro, B):
    p0 = (1-ro)/(1-(ro**(B+1)))
    # N = (1/((1-ro)**2)) - (B*(ro**B)/(1-ro))
    N = ro*(((1-(ro**(B+1))) / ((1-ro)**2)) - ((B+1)*(ro**B)/(1-ro)))
    return (p0*N*arr) - serv
```

```python
df1B = pd.read_excel('mg15.xlsx')
```

```python
plt.plot(df1B["ro"], df1B["pktLoss"]/sent)
plt.plot(df1B["ro"], [p_B(ro, 5) for ro in df1B["ro"]])
```

```python
df = pd.read_excel("mg15.xlsx")
plt.plot(df["ro"], df["avgWaitTime"], label = "from Sim")
plt.plot(df["ro"], [waitTime(df["avgArrTime"][i], df["avgServTime"][i], df["ro"][i], df["B"][i]) for i in range(len(df["ro"]))], label = "calculated", linestyle = 'dashed')
plt.legend()
```

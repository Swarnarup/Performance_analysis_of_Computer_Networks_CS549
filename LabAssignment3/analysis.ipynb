#### Factor:Reading Method (fget & fread) and Factor : File size  Analysis

```python
df0 = pd.read_csv('data.csv', index_col = [0])
df1 = pd.read_csv('data1.csv', index_col = [0])
df2 = pd.read_csv('data2.csv', index_col = [0])
df3 = pd.read_csv('data3.csv', index_col = [0])
```

```python
df = (df0+df1+df2+df3)/4
```

```python
df
```

```python
df.plot(ylabel='Throughput (kb per seconds)', grid=True, title='File Size vs Throughput(kbps) For each methods')
```
